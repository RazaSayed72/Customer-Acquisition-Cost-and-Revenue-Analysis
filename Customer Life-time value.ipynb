```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import warnings 
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("customer_acquisition_data.csv")

#Display all columns
pd.set_option('display.max_columns', None)

# Display the DataFrame
df.head()
```

# 1) Make a Visualization for the distribution of Customer Acquisition Cost.

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x=df['cost'])

# Add labels & title
plt.xlabel("cost")
plt.ylabel("frequency")
plt.title("Customer Acquisition Cost")

# Show plot
plt.show()
```

# 2) Create a Visualization for the revenue generated by the customer.

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x=df['revenue'])

# Add labels & title
plt.xlabel("Revenue")
plt.ylabel("frequency")
plt.title("Distribution of Revenue")

# Show plot
plt.show()
```

# 3) Compare the cost of acquisition across various channels and determine which ones are the most and least profitable?

```python
Cost_of_aqusition = df.groupby('channel')['cost'].sum().reset_index().sort_values(by="channel")
Cost_of_aqusition
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=Cost_of_aqusition, x=df['channel'],y =df['cost'])

# Add labels & title
plt.xlabel("channel")
plt.ylabel("cost")
plt.title("cost of acquisition")

# Show plot
plt.show()
```

```python
# Aggregate sales by Category
Cost_of_aqusition = df.groupby('channel')['cost'].sum().reset_index().sort_values(by="channel")

# Plot Pie Chart
plt.figure(figsize=(8, 6))
plt.pie(
    Cost_of_aqusition["cost"], 
    labels=Cost_of_aqusition["channel"], 
    autopct="%1.1f%%",  
    startangle=140, 
    wedgeprops={"edgecolor": "black"}
)

# Add title
plt.title("cost of acquisition ")
plt.show()
```

# 4) Find out which channels are most and least effective at converting customers?

```python
# Aggregate sales by Category
Channel_conversion = df.groupby('channel')['conversion_rate'].mean().reset_index().sort_values(by="channel")

# Plot Pie Chart
plt.figure(figsize=(8, 6))
plt.pie(
    Channel_conversion["conversion_rate"], 
    labels=Channel_conversion["channel"], 
    autopct="%1.1f%%",  
    startangle=140, 
    wedgeprops={"edgecolor": "black"}
)

# Add title
plt.title("Conversion Rate  by Channel ")
plt.show()
```

# 5) calculate the total revenue by channel and analyze the most and least profit channels in terms of generating revenue.

```python
# Aggregate sales by Category
Revenue_by_channel = df.groupby('channel')['revenue'].sum().reset_index().sort_values(by="channel")

# Plot Pie Chart
plt.figure(figsize=(8, 6))
plt.pie(
    Revenue_by_channel["revenue"], 
    labels=Revenue_by_channel["channel"], 
    autopct="%1.1f%%",  
    startangle=140, 
    wedgeprops={"edgecolor": "black"}
)

# Add title
plt.title("Total Revenue by Channel  ")
```

# 6) calculate the return on investment (ROI) for each channel.

```python
df['ROI']=df['revenue']/df['cost']

# Aggregate sales by Category
Return_on_Investment = df.groupby('channel')['ROI'].sum().reset_index().sort_values(by="channel")

# Plot Pie Chart
plt.figure(figsize=(8, 6))
plt.pie(
    Return_on_Investment["ROI"], 
    labels=Return_on_Investment["channel"], 
    autopct="%1.1f%%",  
    startangle=140, 
    wedgeprops={"edgecolor": "black"}
)

# Add title
plt.title("Total Return on Investment by Channel  ")
```
